```python

#  end year filter

def replace_null(x):
    try:
        return int(x)
    except (ValueError, TypeError):
        return 0  # Replace non-convertible values with 0

# Apply the function to convert "end_year" column
df["end_year"] = df["end_year"].apply(replace_null)

sectors = df["sector"].unique()

list_intensity = []
list_relevance = []

def filter_end_year(df, end_year):

    
    for sector in sectors:
   
        
        date_wise_filtered_data = df[df["end_year"] >= end_year]
        intensity_counts = date_wise_filtered_data[df["sector"] == sector].groupby(df["intensity"]).size()
        relevance_counts = date_wise_filtered_data[df["sector"] == sector].groupby(df["relevance"]).size()
        
        list_intensity.append(date_wise_filtered_data[df["sector"] == sector].groupby(df["intensity"]).size())
        list_relevance.append(date_wise_filtered_data[df["sector"] == sector].groupby(df["relevance"]).size() )
        

filter_end_year(df, 2041)
list_relevance[0]
```
